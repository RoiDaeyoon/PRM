# 5.1 The Distribution Function

<br>
<br>

  *phase space*  
 6차원 공간으로 위치좌표 $x, y, z$와 속도 좌표 $v_{x},v_{y},v_{z}$로 이루어져 있고, 한 시간에서 입자의 역학적 상태는 phase state상에 존재한다.  
입자수가 매우 많을 경우 공간내의 작은 부피내의 입자의 평균수밀도를 정의하는 것은 의미있는 작업이다. 이 밀도는 distribution function $f(r,v,t)$로 불리고 다음과 같이 정의된다.  


### $ dN = f(r,v,t)d^{3}xd^{3}v$</center>

<br>
시간 t에서 phase-space 부피내에 존재할 수 있는 입자의 수를 나타낸다. 입자의 수가 충분히 많아 f가 연속이라고 가정하면 전체 입자수는 dN을 적분함으로 써 얻을 수 있고 다음과 같다.

<br>

### $N = \int^{\infty}_{-\infty} f(r,v,t)d^{3}xd^{3}v$



영역 R에서의 임의의 동역학적 양 $g(r,v)$의 평균은 다음과 같다.


### $<g(r,v)>=\frac{1}{N}\int_{R} g(r,v)f(r,v,t)d^{3}xd^{3}v$

<br>
플라즈마 내에는 수많은 종이 존재하기 때문에  플라즈마의 각 입자에 대해 $f_{s}(r,v,t)$를 정의해야 하고, $g(r,v)$를 적절히 선택함으로써 어려 형태의 평균을 얻을 수 있다. 다음은 주로 쓰이는 평균들이다.


### number density,$n_{s} = \int_{V}f_{s}(r,v,t)d^{3}v$ -5.1.4  
### average velocity, $U_{s} = \frac{1}{n_{s}}\int_{V}Vf_{s}(r,v,t)d^{3}v$  
### kinetic energy densiti, $W_{s} = \int_{V}\frac{1}{2}m_{s}v^{2}f_{s}(r,v,t)d^{3}v$  
### pressure tensor, $\overleftrightarrow{P_{s}}= \int_{V}m_{s}(v-U_{s})(v-U_{s})f_{s}(r,v,t)d^{3}v$

속도 좌표상에서 적분을 하였다. 압력텐서는 $[P_{ij}]$로 j평면에 i방향으로 속도$U_{s}$로 전달되는 모멘텀의 평균rate이다. $(v-U_{s})(v-U_{s})$는 다음과 같은 행렬로 나타난다.

 


$n_{s}와 U_{s}$는 전하밀도 $\rho_{m}$과 질량밀도$\rho_{m}$을 구하는데에 이용된다.  

### $\rho_{q}=\sum_{s}e_{s}n_{s}$  
### $\rho_{m}= \sum_{s}m_{s}n_{s}$

비슷하게 전류 밀도는 다음과 같다.

### $J = \sum_{s}e_{s}n_{s}U_{s}$

위에서 정의된 다양한 평균을 구하는 방법을 보이기 위해 분산함수를 예시를 들어 설명한다.

$v=U_{0}$로 움직이는 전자, $v=0$으로 움직이는 이온의 coldbeam을 고려해보면, 분산함수는 다음과 같다.


### $f_{e}=n_{0}\delta(v-U_{0}) $ and $f_{i}=n_{0}\delta(v)$  

이러한 간단한 경우 5.1.4 식을 사용하여 쉽게 $n_{e}=n_{i}=n_{0}$을 보일 수 있다. 전하 밀도는 $\rho_{q}=0$이다. 전자의 평균속도 $U_{e}=U_{0}$이고 이온의 평균속도는 $U_{i}$이다. 전자는 이온에 대해서 움직이므로 전류는 존재한다. 이 때 전류 밀도는 $J=-en_{0}U_{0}$로 주어진다. 평균속도에 대해 spread가 없으므로 압력텐서 $\overleftrightarrow{P_{e}},\overleftrightarrow{P_{i}}$는 0이다.

다음으로 더 복잡한 맥스웰 속도 분포식을 고려한다.  

### $f=n_{0}(\frac{m}{2\pi\kappa T})^{3/2}exp[-\frac{m(\mathbf{v}-\mathbf{U}_{0})^{2}}{2\kappa T}]$ 

이 분산식은 $V=U_{0}(평균속도)$에서 피크를 갖고 이 속도에서 대칭이다. 압력텐서는 다음과 같다.  

### $P_{ij}=n_{0}\kappa T $ for i=j    
### $P_{ij}=0$ for $i\neq j$


압력텐서는 같은 값을 갖는 원소의 대각행렬이고, 이는 등방성압력을 나타낸다. 압력은 수밀도$n_{0}$에 비례하고 $P_{ij}=n_{0}\kappa T $이므로 온도T에 비례한다.

압력이 이방성인 경우 대각선원소는 같지 않다.  이방성 분산 함수의 예는 다음과 같다. magnetized plasma에서 주고 일어난다.

### $f=n_{0}(\frac{m}{2\pi\kappa T_{\perp}})(\frac{m}{2\pi\kappa T_{\parallel}})^{1/2}exp[-\frac{mv^{2}_{\perp}}{2\pi\kappa T_{\perp}}]exp[-\frac{mv^{2}_{\parallel}}{2\pi\kappa T_{\parallel}}]$ (5.1.16)

$\perp ,\parallel $ 는 자기장으로의 직각과, 평행 방향을 나타낸다. z방향으로 나란하면(aligne) $P_{\perp} = n_{0}\kappa T_{\perp}$ 와 $P_{\parallel}=n_{0}\kappa T_{\parallel}$ 에서 압력텐서는 다음과 같은 간단한 형태가 된다.  

### $\overline{P}=\begin{bmatrix}P_{\perp} & 0 & 0 \\0 & P_{\perp} & 0 \\0 & 0 & P_{\parallel}  \end{bmatrix}$(5.1.17)  


충돌가 없는 조건에서, 평행, 수직인 압력은 분리(decouple)된다.

# 5.2 The Boltzmann and Vlasov Equations

phase space상에서 분산함수의 evolution을 정의하는 미분방정식은 볼츠만 방정식이다. 이 방정식을 유도하기 위해, 입자에 작용하는 힘은 long range, short range 두 타입으로 분류될 수 있다고 가정한다. long range force는 주어진 공간으 모든 입자에 대해 동일하며, short range force는 동일하지 않다.  

long range force는 수 많은 입자들의 상대적으로 긴 거리에서의 collective effect에서 기인하며, short range force는 적은 양의 입자가 상대적으로 가까운 거리에서 기인하고 collision에 해당한다. 충돌는 입자 상호작용의 속도의 크고, 동시적인 변화를 유발한다.

볼츠만 방정식을 유도하기 위해, short range effect(충돌)는 없다고 가정한다. 볼륨 내 모든 입자는 동일한 long range force $F$를 받는다. 따라서 phase space 상에서 입자가 나타내는 힘은 뉴턴 법칙으로 나타난다.  

### $F = m \frac{ dv}{ dt}$  (5.2.1)


이는 절대 교차하지 않는 입자의 궤도를 따른다. 원소들의 evolution은 아래 그림과 같이 나타난다. phase space 볼륨 원소 내의 입자들은 보존되므로 다음과같이 쓸 수 있다. 


### $f(r,v,t)d^{3}xd^{3}v = f'(r',v',t')d^{3}x'd^{3}v'$ (5.2.2)
f는 t에서의 분산함수, f'는 t+dt에서의 분산함수이다.

phase space에서의 입자의 모션식은 다음과 같이 주어진다.  

### $x'_{r} = x_{r}+v_{r}dt$ and $v'_{r}=v'_{r} + \frac{F_{r}}{m}dt$  (5.2.3)



$x_{r},v_{r}$을 나타내기 위해 인덱스 표기 $r = (1, 2, 3)$을 사용한다. 미소 부피내에서 $(r,v) \rightarrow (r',v')$ 변환은 다음과 같이 결정된다.  

### $d^{3}x'd^{3}v' = Jd^{3}xd^{3}v$  (5.2.4)


여섯개의 변형 방정식이 존재하므로 자코비안은 6X6이다.  



적절한 미분값을 구하고, dt항을 정리하면 다음과 같이 쓸 수 있다.


### $d^{3}x'd^{3}v' = [1+\sum_r^{}\frac{\partial v_{r}}{\partial x_{r}}dt + \sum_r^{}\frac{\partial }{\partial v_{r}}(\frac{F_{r}}{m})dt]d^{3}xd^{3}v$(5.2.6)


미분을 완료하기 위해 식 5.2.2의 $f'$를 chain rule을 사용해 테일러 전개한다. $v_{r}=dx_{r}/dt$ 와 $dv_{r}/dt=F_{r}/m$을 사용하면 다음과 같은 식을 얻는다.

### $f' = f+\frac{\partial f}{\partial t}dt +\sum_r^{}v_{r}\frac{\partial f}{\partial x_{r}}dt + \sum_r^{}(\frac{F_{r}}{m})\frac{\partial f}{\partial v_{r}}dt$ (5.2.7)

5.2.2에 5.2.6의 $d^{3}x'd^{3}v'$과 5.2.7의 $f'$을 빼면 5.2.8을 얻는다.

### $f = f +\frac{\partial f}{\partial t}dt + \sum_r^{}v_{r}\frac{\partial f}{\partial x_{r}}dt+\sum_r^{}(\frac{F_{r}}{m})\frac{\partial f}{\partial v_{r}}dt+\sum_r^{}f\frac{\partial v_{r}}{\partial x_{r}}dt+\sum_r^{}f\frac{\partial }{\partial v_{r}}(\frac{F_{r}}{m})dt$  (5.2.8)

1차 편미분 방정식으로 간소화된다.

### $\frac{\partial f}{\partial t}+\sum_r^{}\frac{\partial }{\partial v_{r}}(v_{r}f)+\sum_r^{}\frac{\partial }{\partial v_{r}}(\frac{F_{r}}{m}f)=0$ (5.2.9)

위 식을 3차원 mass continuity 방정식과 비교하면 6차원 countinuity 방정식으로 볼 수 있다.  

### $\frac{\partial \rho_{m}}{\partial t}+\sum_r^{}\frac{\partial }{\partial x_{r}}(v_{r} \rho_{m})=0$(5.2.10)

식 5.2.9는 상대속도, 비 상대속도 모두에 유효하다. 두 개의 단순호가 가능하다.

1. $v_{r},x_{r}$은 독립변수이고, 이는 $\frac{\partial{v_{r}}}{\partial{x_{r}}}=0$ 을 의미한다. $v_{t}$는 $x_{r}$의 독립성은 한눈에 명확하지는 않지만, 입자 mechanics $v(t)$는 시간 미분$r(t)$로 부터 얻어지므로, $r(t)$에 의존한다. 그러나 속도론적 정의는 phase space에서의 개별 입자에 기초하지 않는다. 이는,어느 시간의 부피 원소에서의 수많은 입자의 평균에 기초한다. 공간좌표의 부피원소는 속도에 독립적으로 선택될 수 있으므로, $x_{r}, v_{r}$은 독립변수이다. 

2. 우리가 mass가 속도에 독립적인 non-relativistic 속도만을 고려하면, $(\partial/\partial_{r})(F_{r}/m)=0$이다. 이 결과는 적절한 미분을 통해 전자기력통해 입증될 수 있다.

r=x에서 다음을 따른다


### $\frac{\partial }{\partial v_{x}}(\frac{F_{x}}{m})=\frac{q}{m}\frac{\partial }{\partial v_{x}}[E_{x}+v_{y}B_{z}-v_{z}B_{y}]=0$  (5.2.11)

r=y, r=z 에서도 비슷한 결과를 얻는다. 이 두 가지 단순화는 J=1을 의미한다. phase-space 흐름은 부피보존이다. 즉, $d^{3}x'd^{3}v'=d^{3}xd^{3}v$를 의미한다. 이러한 단순화를 통해 식 5.2.9는 다음과 같이 나타난다. 


### $\frac{\partial f}{\partial t}+\sum_r^{}v_{r}\frac{\partial f}{\partial x_{r}}+\sum_r^{}\frac{F_{r}}{m}\frac{\partial f}{\partial v_{r}}=0$ (5.2.12)

혹은 벡터 표기로

### $\frac{\partial f}{\partial t}+ \mathbf{v}\cdot\triangledown f + \frac{\mathbf{F}}{m} \cdot\triangledown_{\mathbf{v}}f=0$(5.2.13)

으로 나타난다.

$\Delta_{v}$는 속도 공간의 그래디언트연산자이고, $\Delta$ 이다. 5.2.11의 질량이 속도에 의존하기 때문에 결과는 relativistic 속도를 hold하지 못한다. $m_{0}$  


다음으로, short-range 힘을 고려한다. short-range 힘은 부피 원소내 모든 입자에서 동일하지 않으므로, 충돌은 입자를 phase-space 내외롤 분산시키는 역할을 한다. 앞서 말했듯, short-range 상호작용의 주 효과는 입자속도의 거의 불연속적인 변화를 야기한다. phase-space의 몇몇 일반적인 경로는 아래 그림에 나타나 있다.  




시간 t에서 t+dt로의 phase-space 부피원소의 성장과 같은 충돌에 의해 분산함수의 변화는 충돌 연산자 $\delta_{c}f/\delta t$를 추가함으로써 고려한다. 충돌연산자는 식 5.2.13의 우측항으로, 아직 정의되지 않았따.

### $\frac{\partial f}{\partial t}+ \mathbf{v}\cdot\triangledown f + \frac{\mathbf{F}}{m} \cdot\triangledown_{\mathbf{v}}f=\frac{\delta_{c}f}{\delta t}$(5.2.14)

이 방정식이 볼츠만 방정식이다. 볼츠만 방정식의 왼쪽항은 입자의 long-range collective 힘을 나타내고, 우측항은 short-range 붕괴 힘을 나타낸다. 

앞서 논의되었던 것처럼 디바이큐브당 전자 수가 크다면 $n\lambda^{3}_{D}>>1$, 그리고 collective 힘이 충돌 힘보다 매우 더 중요하다. 이러한 조건에서 충돌텀은 무시할 수 있다. Lorentz force와 함께 결과 방정식은 다음과 같다.  

### $\frac{\partial f}{\partial t} + \mathbf{v}\cdot\triangledown f + \frac{q}{m}[\mathbf{E}+\mathbf{v}\times\mathbf{B}]\cdot\triangledown_{v}f=0$ (5.2.15)


이 방정식은 Vlasov 방정식이라고 불린다.

# 5.3 Solution Based on Constants of the Motion


만약 계가 하나 이상의 모션 상수를 알고있다면, 즉시 Vlasov 방정식의 해를 쓰는 것이 가능하다. 이러한 해들은 분산함수가 phase space에서의 representative 궤도를 따라 분산함수가 일정하다는 사실에 근거한다. 이 결과를 증명하기 위해, f를 representative 궤도 $r(t)와 v(t)$를 따라 미분을 한다. 체인룰을 사용하면 다음과 같다.


### $df_{s}=\frac{\partial f_{s}}{\partial t}dt + \sum_r^{}\frac{\partial f_{s}}{\partial x_{r}}\frac{dx_{r}}{dt}dt +\sum_r^{}\frac{\partial f_{s}}{\partial v_{r}}\frac{dv_{r}}{dt}dt$ (5.3.1)

$d_{x}/dt=v_{r}$과 $dv_{r}/dt=F_{r}/m$을 알고, dt로 나누면 다음을 얻는다.  


### $\frac{df_{s}}{dt}=\frac{\partial f_{s}}{\partial t}+\mathbf{v}\cdot\triangledown f_{s}+\frac{\mathbf{F}}{m}\cdot\triangledown f_{s}=0$ (5.3.2)


우측항은 Vlasov 방정식이므로 0이다. $df_{s}/dt=0$이므로 $f_{s}$가 궤도에 따라 상수임을 얻는다. 가능한 궤도는 모션상수$C_{m}(r,v)$에 의하여 제한된다. 모션 상수의 함수들은 다음과 같이 표현되며 Vlasov 방정식의 해이다.


### $f_{x}[C_{1}(\mathbf{r},\mathbf{v}),C_{2}(\mathbf{r},\mathbf{v}),...]$ (5.3.3)

이 절차의 예시로, 정전기적 포텐셜 $\Phi(r)$ 전하 q, 질량 m 인 입자의 움직임을 생각해보자. 이 계에서 전체 에너지는 다음과 같고, 모션 상수이다.


### $W=\frac{1}{2}mv^{2}+q\phi(\mathbf{r})$ (5.3.4)

$f_{s}[W(r,v)]$로 나타나는 모든 식은 Vlasov 방정식의 해이다. 정확한 해의 형태는 경계조건에 의해 정해진다. 예를들어, $f(r,v)$ $\Phi(\infty)$인 $r\rightarrow \infty$ 인때의 $T_{0}$, 맥스웰 수밀도$n_{0}$로 알려져있다면 어떠한 임의의 점에서의 일반 해는 다음과 같다.


### $f(\mathbf{r},\mathbf{v})=n_{0}(\frac{m}{2\pi\kappa T_{0}})^{3/2} e^{-[\frac{1}{2}mv^{2}+q\phi(\mathbf{r})]/\kappa T_{0}}$ (5.3.5)

이 분산함수는 2.2.2 식에서 점전하 주위의 plasma sheath에서의 포텐셜을 결정하는데에 사용됐던 것과 recognized된다. 


모션 상수를 사용하여 Vlasov 방정식의 해를 구할 때, 경계조건에 주의해야한다. 만약 정전기적 포텐셜이 아래 그림과 같이 단조롭다면, phase space (x, v)의 한 지역에서 경계점 $x_{0}$로 부터 접글할 수 없다. 접근할 수 없는 지역의 예시는 cross-hatching에 의해 보여진다.




cross-hatched 지역의 궤도는 $x_{0}$를 통과할 수 없으므로, $x_{0}$에서의 $f$의 경계조건은 분산함수를 명시하는데 필요하다. 만약 접근불가능 지역에서 소스가 주어지지 않으면, 그 지역에서 분산함수는 0이다. 실제로는 충돌과 다양한 wave입자 상호작용은 종종 입자들은 접근 불가능한 지역으로 분산시킨다. 이러한 상황에서는 정확한 분산함수를 결정하는 과정이 어렵고, 이러한 scattering 효과를 고려해야한다.

# 5.4 The Moment Equations


플라즈마 역학에서 우리는 분산함수보다 밀도, 압력같은 거시적인 평균에 더 관심이 있다. 이는 모멘텀 방정식을 쓰면 간편하다. 모멘텀 방정식은 볼츠만 방정식에 속도의 지수승을 곱한 후 속도 공간에서 적분하여 구해진다. 이 절차는 각 승수마다 하나의 모멘텀식을 생성한다. 첫 세개의 모멘텀 방정식을 소개한다.

# 5.4.1 Zeroth Moment

0차 모멘트를 얻기 위해, 우리는 볼츠만 방정식에 $v^{0}=1$을 곱하고 속도 공간에서 적분하면 다음을 얻는다.

### $\int_{V}^{}\frac{\partial f_{s}}{\partial t}d^{3}v + \int_{V}^{}\mathbf{v}\cdot\triangledown f_{s}d^{3}v+\frac{1}{m_{s}}\int_{V}^{} \mathbf{F}\cdot\triangledown_{\mathbf{v}}f_{s}d^{3}v = \int_{V}^{} \frac{\delta_{c}f_{s}}{\delta t}d^{3}v$ (5.4.1)


이 방정식의 왼쪽의 첫 번째 적분은 다음과 같다. $n_{s}$는 평균 수밀도이다.

### $\int_{V}^{}\frac{\partial f_{s}}{\partial t}d^{3}v=\frac{\partial }{\partial t}\int_{V}^{}f_{s}d^{3}v= \frac{\partial n_{s}}{\partial t}$ (5.4.2)


비슷하게, 두 번째 적분은 다음과 같다. $U_{s}$는 평균 속력이다.

### $\int_{V} v\cdot \triangledown f_{s}d^{3}v = \triangledown \cdot\int_{V} vf_{s}d^{3}v = \triangledown \cdot (n_{s}U_{s})$ (5.4.3)

세 번째 적분은 무한대에서 부피적분에서 면적분으로의 변환을 위해 사용된 가우스 이론으로 다시 쓸 수 있다.

### $\int_{V} F\cdot \triangledown_{v} f_{s}d^{3}v =\int_{V} \triangledown_{v}\cdot(Ff_{s})d^{3}v =\int_{S}(Ff_{s})\cdot ds_{v}d^{3}v$(5.4.4)


표면 적분은 $f(v)$가 무한대 속도에서 춘분히 빠르게 감소하면 없는 것으로 가정한다. 충돌은 밀도를 변화시키지 않는다고 가정하기 때문에 충돌항을 갖는 5.4.1의 우측항의 적분은 0이다.

### $\int_{V}\frac{\delta_{c}f_{s}}{\delta t}d^{3}v = \frac{\delta_{c}n_{s}}{\delta t}$(5.4.5)


5.4.2 ~ 5.4.5를 합치면, 0차 모먼트는 다음과 같다.


### $\frac{\partial n_{s}}{\partial t}+\nabla\cdot (n_s\mathbf{U_s})=0$ (5.4.6)


이 방정식은 간단하게 s번째 종의 연속 방정식이다. 위 식에 전하 $e_{s}$를 곱하고 모든 종을 더하면 전하 연속 방정식을 얻는다. 이는 전기, 자기 방정식의 기초이다.

### $\frac{\partial \rho_q}{\partial t}+ \nabla \cdot\mathbf{J}=0$ (5.4.7)

# 5.4.2 First Moment

1차 모멘트는 5.2.14 볼츠만 방정식에 $v$를 곱하고 속도 공간에 대해 적분하여 얻는다. 이는 모멘텀 $m_{s}v$를 곱하는것 보다 더 편리하다.  

### $m_s\int_{V}\mathbf{v}\frac{\partial f_s}{\partial t}d^3v+m_s\int_{V}\mathbf{v}(\mathbf{v} \cdot \nabla f_s)d^3v + \int_{V} \mathbf{v}(\mathbf{F}\cdot \nabla_{\mathbf{v}} f_s)d^3v = m_s\int_{V} \mathbf{v}\frac{\delta_c f_s}{\delta t}d^3v$ (5.4.8)


왼쪽 첫 번째 적분은 다음과 같이 쓰일 수 있다.  

### $m_s\int_{V} \mathbf{v} \frac{\partial f_s}{\partial t}d^3v=\frac{\partial}{\partial t}\left ( m_s \int_{V} \mathbf{v} f_sd^3v\right )=\frac{\partial}{\partial t}(m_sn_s\mathbf{U_s})$ (5.4.9)


두 번째 적분은 다음과 같다.

### $m_s\int_{V}\mathbf{v}(\mathbf{v} \cdot \nabla f_s)d^3v=\nabla \cdot \left [ m_s\int_{V} \mathbf{v} \mathbf{v} f_s d^3v \right ]$ (5.4.10)

다음 관계를 이용하여 $vv$텐서를 편리하게 다시 쓸 수 있다.

### $\mathbf{v} \mathbf{v} =(\mathbf{v}-\mathbf{U_s})(\mathbf{v}-\mathbf{U_s})+\mathbf{U_s}\mathbf{v} +\mathbf{v}\mathbf{U_s} -\mathbf{U_s}\mathbf{U_s}$ (5.4.11)

따라서 5.4.10의 적분항은 다음과 같이 간단하게 나타내진다.


### $m_s\int_V\mathbf{v}\mathbf{v}f_sd^3v =m_s\int_V(\mathbf{v}-\mathbf{U_s})(\mathbf{v}-\mathbf{U_s})f_sd^3v+m_sn_s\mathbf{U_s}\mathbf{U_s}$ (5.4.12)


위 방정식의 우측 항의 첫 번째 텀은 5.1.7식에 의해 정의된 단순 압력항이다. 식 5.4.10은 다음과 같이 다시 쓰일 수 있다.

### $m_s\int_V\mathbf{v}(\mathbf{v} \cdot \nabla f_s)d^3v =\nabla \cdot \mathbf{\overleftrightarrow{P_S}}+\nabla \cdot (m_sn_s\mathbf{U_s}\mathbf{U_s})$ (5.4.13)

왼쪽의 세 번째 적분항은 다음과 같다.

### $\int_{V} \mathbf{v}(\mathbf{F}\cdot \nabla_{\mathbf{v}} f_s)d^3v = \int_V (v_x \mathbf{\hat{x}}+v_y \mathbf{\hat{y}}+v_z \mathbf{\hat{z}})\left [ \frac{\partial}{\partial v_x}(F_xf_s)+ \frac{\partial}{\partial v_y}(F_yf_s) \frac{\partial}{\partial v_z}(F_zf_s)\right ]d^3v$ (5.4.14)

Lorentz 힘을 이용하였다.
### $\mathbf{F}\nabla_\mathbf{v}f_s=\nabla_\mathbf{v} \cdot (\mathbf{F}f_x)$ (5.4.15)

반드시 계산되어야 하는 9개의 적분 중 6개의 타입은  


###  $\int_V v_y\frac{\partial}{\partial v_z}(F_zf_s)d^3v = \int_{-\infty }^{\infty}\int_{-\infty }^{\infty}dv_xdv_yv_y\int_{-\infty }^{\infty}\frac{\partial}{\partial v_z}(F_zf_s)dv_z$ (5.4.16)  

이고, 적분하여 다음과 같은 식을 제공한다.


### $\int_{-\infty }^{\infty}\frac{\partial}{\partial v_z}(F_zf_s)dv_z=[F_zf_s]^{\infty}_{-\infty}$ (5.4.17)


$f_{s}$가 무한대의 속도에서 충분히 빠르게 0으로 간다면, 이러한 적분항은 모두 0이다. 남아있는 세 개의 적분항은 다음과 같다..

### $\int_{-\infty }^{\infty}\int_{-\infty }^{\infty}dv_xdv_y\int_{-\infty }^{\infty}dv_zv_z\frac {\partial}{\partial v_z}(F_zf_s)$ (5.4.18)


적분하면 다음과 같다. 

### $\int_{-\infty }^{\infty}dv_zv_z\frac{\partial}{\partial v_z}(F_zf_s)=[v_zF_zf_s]^{\infty}_{-\infty}-\int_{-\infty}^{\infty}F_zf_sdv_z$ (5.4.19)

한번 더 $f_{s}$가 빠르게 0으로 수렴한다고 가정하면, 우측의 첫 번째 항은 0이다.이러한 적분항을 모두 조합하면, 5.4.14는 다음과 같이 쓰일 수 있다.

### $\int_{V} \mathbf{v}(\mathbf{F}\cdot \nabla_{\mathbf{v}} f_s)d^3v=-\int_V\mathbf{F}f_sd^3v$ (5.4.20)

Lorentz 힘 F를 빼준 후 적분하면, 위의 식은 다음과 같은 간단한 결과를 낳는다.


### $\int_{V} \mathbf{v}(\mathbf{F}\cdot \nabla_{\mathbf{v}} f_s)d^3v=-e_sn_s(\mathbf{E}+\mathbf{U_s}\times \mathbf{B})$ (5.4.21)

마지막으로, 5.4.7의 우측 충돌항은 다음과 같이 쓸 수 있다.

### $m_s\int_V\mathbf{v}\frac{\delta_cf_s}{\delta t}d^3v=\frac{\delta_c \mathbf{p_s}}{\delta t}$ (5.4.22)

$\delta_c \mathbf{p_s}/\delta t$는 충돌로 인한 부피당 평균 모멘텀 변화율이다. 이 항은 충돌로 인한 항력을 나타내며, 5.6절에서 다룬다.

위 식들을 조합하여 다음 식을 얻는다.


### $\frac{\partial}{\partial t}(m_sn_s\mathbf{U_s})+\nabla \cdot (m_sn_s\mathbf{U_s}\mathbf{U_s})=n_se_s(\mathbf{E}+\mathbf{U_s}\times {\mathbf{B}})-\nabla \cdot \mathbf{\overleftrightarrow{P_s}}+\frac{\delta_c \mathbf{p_s}}{\delta t}$ (5.4.23)


왼쪽의 두 항은 시간 미분항을 확장시키고, $\nabla \cdot(\mathbf{AB})=(\mathbf{\nabla \cdot A})\mathbf{B}+(\mathbf{A \cdot \nabla})\mathbf{B}$를 사용하여 더 유용한 식으로 쓸 수 있다.


### $\frac{\partial}{\partial t}(m_sn_s\mathbf{U_s})+\nabla \cdot(m_sn_s\mathbf{U_s}\mathbf{U_s})=m_sn_s\frac{\partial \mathbf{U_s}}{\partial t}+m_s\mathbf{U_s}\left [ \frac{\partial n_s}{\partial t}+\nabla \cdot (n_s\mathbf{U_s}) \right ]+m_sn_s(\mathbf{U_s}\cdot \nabla)\mathbf{U_s}$ (5.4.24)


5.4.6의 연속 방정식에 의해, 괄호친  항은 0이다. 따라서 볼츠만 방정식의 1차 모멘트는 다음과 같다.

### $m_sn_s\left [ \frac{\partial \mathbf{U_s}}{\partial t}+(\mathbf{U_s} \cdot \nabla)\mathbf{U_s} \right ]=n_se_s[\mathbf{E}+\mathbf{U_s}\times \mathbf{B}]-\nabla \cdot \overleftrightarrow{\mathbf{P_s}}+\frac{\delta_c \mathbf{p_s}}{\delta t}
$ (5.4.25)

왼쪽의 괄호 항은 유체 역학에서 자주 일어나며, 대류 미분항이라고 불린다. $\partial U_{s}/\partial t$항은 분명한 시간 변화에 따른 속도 변화율을 제공하고, $(U_{s}\dot \Delta)U_{s}$는 공간적 변화에 따른 속도 변화율을 제공한다. 우리가 속도 $U_{s}$로 이동하는 부피 원소를 따르면, 대류 미분항은 간단히 상미분항으로 변하고, 유체 원소의 궤적을 따라 evaluate한다.

### $\frac{d\mathbf{U_s}}{dt}= \frac{\partial \mathbf{U_s}}{\partial t}+(\mathbf{U_s} \cdot \triangledown)\mathbf{U_s}$ (5.4.26)


5.4.25는 다음과 같이 쓸 수 있다.

### $m_sn_s \frac{d \mathbf{U_s}}{dt} = n_s e_s [\mathbf{E}+\mathbf{U_s}\times \mathbf{B}]-\triangledown \cdot \overleftrightarrow{\mathbf{P_s}}+\frac{\delta _c \mathbf{p_s}}{\delta t}$ (5.4.27)

이 식은 모멘텀 방정식이라고 불리며, s 번째 원소의 플라즈마에서의 뉴턴 법칙이라고 여겨진다. 왼쪽항은 마치 입자인것처럼 부피 원소에서 evaluate 된 단위 부피당 모멘텀의 변화율을 나타낸다. 우측 첫 항은 단위부피당 전자기력, 두 번째항은 압력 구배에 의한 단위 부피당 힘, 세 번째 항은 단위 부피당 충돌로인한 항력이다.

# 5.4.3 Second Moment


2차 모멘트는 볼츠만 방정식에 텐서 $vv$를 곱하고 속도 공간에서 적분함으로써 얻을 수 있다. 이는 9개의 원소를 가진 대칭 행렬을 얻게 되고, 대칭행렬의 비 대각원소는 서로 같으므로 6개는 독립적이다. 운동에너지 $(1/2)m_{s}v^{2}$을 볼츠만 방정식에 곱하고 적분하면 다음과 같은 식을 얻는다.

### $ \frac{m_s}{2}\int_V \nu^2 \frac{\partial f_s}{\partial t}\mathrm{d^3}\nu +\frac{m_s}{2}\int_V \nu^2 (\mathbf{v}\cdot \triangledown f_s)\mathrm{d^3}\nu+\frac{1}{2}\int_V \nu ^2 (\mathbf{F}\cdot \mathbf{\triangledown_v}f_s)\mathrm{d^3}\nu = \frac{m_s}{2}\int_V \nu^2 \frac{\delta _c f_s}{\delta t}\mathrm{d^3}\nu $ (5.4.28)


왼쪽항 첫 번째 적분항은 다음과 같이 쓸 수 있다.

### $\frac{m_s}{2}\int_V \nu^2 \frac{\partial f_s}{\partial t}\mathrm{d^3}\nu = \frac{\partial}{\partial t} \left ( \int_V \frac{1}{2}m_s \nu^2 f_s \mathrm{d^3}\nu \right )=\frac{\partial W_s}{\partial t} $ (5.4.29)

$w_{s}$는 운동에너지 밀도이다. 두 번째 적분항은 다음과 같다.


### $\frac{m_s}{2}\int_V \nu^2 (\mathbf{v} \cdot \triangledown f_s)\mathrm{d^3}\nu =\triangledown \cdot \left ( \int_V \frac{1}{2}m_s \mathbf{v}\nu^2 f_s \mathrm{d^3}\nu \right)= \triangledown \cdot \mathbf{Q}_s$ (5.4.30)

### $ \mathbf{Q}_s = \int _V \frac{1}{2}m_s \nu ^2 \mathbf{v} f_s \mathrm{d^3}\nu $ (5.4.31)
은 운동에너지 플럭스이다. 세 번째 적분항은 무한대 속도에서의 기여를 버린 후 적분하여 단순화한다.  

### $ \frac{1}{2}\int_V \nu^2 (\mathbf{F}\cdot \triangledown _\mathbf{v} f_s) \mathrm{d^3}\nu = -\frac{1}{2}\int_V f_s (\mathbf{F}\cdot \triangledown _\mathbf{v} \nu^2) \mathrm{d^3}\nu $ (5.4.32)

로렌츠 힘, $\triangledown_{v}v^{2}=2v$, $v\cdot (v\times B)$를 통해 윗 식의 우측 항은 다음과 같이 나타낼 수 있다.

### $ \frac{1}{2}\int_V \nu^2 (\mathbf{F}\cdot \triangledown _\mathbf{v} f_s) \mathrm{d^3}\nu =   - e_s \mathbf{E}\cdot \int_V \mathbf{v}f_s \mathrm{d^3}\nu = -\mathbf{E} \cdot (n_s e_s \mathbf{U}_s) = -\mathbf{E \cdot J}_s $ (5.4.33)


$J_{s}$는 s 종에 의해 현재 기여하고 있다. 위 식들을 이용하여 나타낸 2차 모멘트는 다음과 같다.

### $\frac {\partial W_s}{\partial t} + \triangledown \cdot \mathbf{Q}_s -\mathbf{E \cdot J}_s = \int_V \frac{1}{2}m_s \nu^2 \frac{\delta_c f_s}{\delta t}\mathrm{d^3}\nu $ (5.4.34)

이 방정식은 s번쩨 종의 에너지 보존 상태를 나타내므로 에너지 방정식이라고 불린다. $E\cdot J_{s}$는 s번째 종의 joule heating에 기여하고, 우측의 충돌항은 s번째 종이 다른 종과 충돌하여 발생하는 에너지 전달률을 나타낸다. 윗 식을 모든 종에대해 더하면 플라즈마 내 전체 에너지 보존 방정식은 다음과 같다.

### $\frac{\partial W}{\partial t} + \triangledown \cdot \mathbf{Q} = \mathbf{E \cdot J} $ (5.4.35)

W는 전체 운동에너지이고, Q는 전체 운동에너지 플럭스, $E\cdot J$는 줄 히팅이다. 완전히 이온화된 플라즈마에서, 전체 에너지의 변화 없이 한 종에서 다른 종으로 쉽게 에너지가 전달되므로 충돌항의 합은 0이다.

# 5.4.4 The Closure Problem 


어떠한 유한한 수의 모멘텀 방정식이라도, 알지 못하는 모멘텀의 수는 항상 방정식의 수를 초과한다. 모르는 모멘트를 풀 수 있는 충분한 방정식은 절대 없다. 이를 *Closure Problem*이라고 부른다.


방정식의 Colsed set을 얻기 위해 (n+1)번 째 모멘트를 n번째 모멘트에 관한 식으로 나타내야 한다. 예를들어, 연속방정식과 모멘텀 방정식만을 사용할때, 방정식의 닫힌 계는 압력텐서, 수밀도와 관련된 상태 방정식을 명시해야 얻을 수 있다. 불행히도, 이는 무작위 충돌을 하는 플라즈마에서는 불가능하다.


일반적으로 상태방정식은 열역학에서 사용되는 잘 알려진 상태방정식에 기초하여 가정한다. 이러한 방식은 국부 열역학적 평형 상태의 플라즈마를 유지하기 위해 충분한 충돌이 있는 플라즈마에서는 잘 맞지만, 충돌이 없는 플라즈마에서는 종종 실패한다. 주로 사용되는 상태방정식을 소개한다.

## <center> The Cold Plasma Equation of State</center>

온도가 매우 낮아, 압력을 무시할 수 있다고 가정한다. 어떠한 $n_{s}$에서도 $\overleftrightarrow{P}=0$이다. 모멘텀 방정식은 다음과 같아진다.

### $ m_sn_s\frac{\mathrm{d}\mathbf{U}_s}{\mathrm{d}t}=n_s e_s[\mathbf{E}+\mathbf{U}_s\times \mathbf{B}]+ \frac{\delta _c \mathbf{p}_s}{\delta t} $ (5.4.36)

만약 충돌 항력항이 0으로 가정되면, 수밀도는 식에서 제외할 수 있다. 모멘텀 방정식은 단일 입자를 위한 모션 방정식으로 변한다. 

### $ m_s\frac{\mathrm{d}\mathbf{U}_s}{\mathrm{d}t}= e_s[\mathbf{E}+\mathbf{U}_s\times \mathbf{B}] $ (5.4.37)


$U_{s}$는 유체 속도 장으로 r, t 좌표에 독립적인 함수이다. 일부 경우에, 대류 보존항 $dU_{s}/dt$는 두 항을 포함하고 있으므로 중요하다. 


### $ \frac{\mathrm{d}\mathbf{U}_s}{\mathrm{d}t}= \frac{\partial \mathbf{U}_s}{\partial t}+ (\mathbf{U}_s \cdot \triangledown)\mathbf{U}_s $ (5.4.38)

유체가 작은 진폭의 파인 경우, 우측의 두 번째 항은 무시된다. 반면에 유체가 움직이는 경우, 연산자 $dU_{s1}/dt$의 정확한 선형화 형태는 다음과 같다.  


### $ \frac{\mathrm{d}\mathbf{U}_{s1}}{\mathrm{d}t}= \frac{\partial \mathbf{U}_{s1}}{\partial t}+ (\mathbf{U}_{s0} \cdot \triangledown)\mathbf{U}_{s1} $ (5.4.39)


퓨리에 변환을 하면 다음과 같은 식이 된다.

### $(-\mathrm{i}\omega ')\tilde{\mathbf{U}}_{s1}=(-\mathrm{i}\omega+\mathrm{i}\mathbf{k}\cdot \mathbf{U}_{s0})\tilde{\mathbf{U}}_{s1}$  (5.4.40)

$\omega '$는 입자와 reference moving의 frame 에서의 진동수이다.

## <center>The Adiabatic Equation of State</center>


단열 상태방정식의 개념은 열역학에서 부터 왔다. 이상기체 상태방정식 $PV = N\kappa T$, 열역학 제 1법칙 $dQ = dU + PdV$, 열흐름이 없으면, $PV^{\gamma}=Constant$에 의해 P,V가 주어진다. 단열 상태방정식은 가스 압축이 너무 빨라서 열이 흐를 시간이 없는 상태에 적용된다. 


음파나 빠르게 움직이는 피스톤에 의한 가스의 압축은 일반적인 단열 상태방정식의 예시이다. 비록 열흐름이 무시되기 위해 압축이 충분히 빠르게 일어나야 하지만, 그렇게 빨리 일어나지 않으므로 국부 열적 평형상태를 유지할 수 없다. 예를들어, 단열 상태방정식은 evacuated 상자내의 자유팽창을 묘사할 수 없다.

열적 평형은 충돌에 의해 유지되기 때문에, 충돌없는 플라즈마에 단열 상태방정식이 적용될 수 있는지 의문이 생긴다. 명백한 답은 '아니다' 이다. 그러나, microscopic wave- 입자 상호작용은 때때로 일반 가스에서의 붕괴와 비슷한 역할을 한다. 단열 충돌없는 플라즈마에서도 상태방정식은 적절한 근사를 제공한다.

단열 상태방정식이 사용될 때 충돌이 isotropic 속도 분산을 유지한다고 가정한다. 이러한 조건 하, 압력 텐서는 같은 원소를 가지는 대각행렬이 된다.

### $\overleftrightarrow{\mathbf{P}}_s=\overleftrightarrow{\mathbf{1}}P_s$ (5.4.41)

$\overleftrightarrow{\mathbf{1}}$는 단위 텐서다. 일반적으로 수밀도를 독립변수로 많이 쓰므로, 단열 상태방정식은 다음과 같이 다시쓸 수 있다.

### $ \frac{\mathrm{d}}{\mathrm{d}t}(P_s/n^\gamma_s)=0 $ (5.4.42)


압력은 이상기체 상태방정식을 따른다. $\gamma$는 자유도 수 f에 의해 정해진다.


### $\gamma = \frac{f+2}{f}$ (5.4.43)


충돌률이 너무 낮아서 isotropic 분산을 유지할 수 없으면 압축 공정의 기하학적인 구조로 자유도 수를 조절한다. 




## <center>The Chew-Goldberger-Low Equation of State</center>


자기 플라즈마 내 충돌률이 너무 작아 평행, 수직 방향으로 모멘텀을 효율적으로 전달할 수 없으면, isotropic 속도 분산은 유지될 수 없다. 자기장에 평형, 수직인 압력들이 같지 않은 경우에서의 변형된 단열 상태방정식이 개발되었다. 자기장은 z축에 ㅕㅇ행하고, 압력텐서는 대각이고 다음과 같이 나타난다.


### $\overleftrightarrow{\mathbf{P}}_s = \begin{vmatrix}
P_{s\perp} & 0 & 0 \\ 
0 & P_{s\perp} & 0\\ 
0 & 0 & P_{s\parallel}
\end{vmatrix} $ (5.4.44)


수직, 평행 압력은 자기장과 수밀도에 관련되어 있다.

### $\frac{\mathrm{d}}{\mathrm{d}t}\left ( \frac{P_{s\perp}}{n_sB}  \right ) =0 \ \mathrm{and} \ \frac{\mathrm{d}}{\mathrm{d}t}\left ( \frac{P_{s\parallel}B^2}{n_s^3}  \right )=0$ (5.4.45)

식 5.1.16으로 주어진 맥스웰 속도 분포는 식 5.4.44로 나타난 압력 텐서를 가진다.
