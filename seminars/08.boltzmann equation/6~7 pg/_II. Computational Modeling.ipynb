## Introduction

　　약간의 구성성분이 존재하는 모든 시스템의 보조 속도 벡터뿐만 아니라 DBD시스템의 모든 개별 입자의 위치를 추적하는 것은 매우 어렵다. 유일한 정확한 해답은 각각의 구성성분을 미세한 스케일에서 추적하는 것이다. 하지만 계산시간의 제한과 메모리 요구량으로 인해 근사해를 얻을 수 밖에 없다.  
        
        
  물리적으로 의미있는 해를 얻기 위해, 볼츠만 방정식을 통해 DBD 계를 지배하는 방적식을 유도해야한다. 이 방정식은 다양한 상태, 형상의 수치적 모델링이 가능해야하고 실제 모델의 물리적 성질과 너무 많은 차이가 나면 안된다.  
  
  따라서, 계 내의 전자와 양이온의 이동과 위치를 추적하기위해 거시적 유체 방정식이 사용되고, 각각의 잘린 공간에서 국부적인 평균 전자 에너지 밀도가 사용된다.

#### <center> 볼츠만 방정식의 정의</center>



이 방정식은 분산함수 f, (혹은 특정 속도($\vec{\nu}$)와 위치에서의 플라즈마 종을 찾을 확률)의 시간적 변화를 나타낸다.  

- 첫 번째 항
 
 시간에 따른 분산 함수의 편화를 의미한다. 이 항이 궁극적으로 주어진 초기조건으로 모든 시간에 대해 풀게될 항이다.  


- 두 번째 항
 
 입자가 특정 속도와 방향으로 계를 통해 움직이는 동안의 공간, 혹은 배치 좌표상에서의 변화를 나타낸다.  


- 세 번째 항
 
 charge 된 입자가 외력으로 인해 가속될 때 플라즈마 종의 분산이 어떻게 변하는지를 보여준다.


- 마지막 우측 항
 
 시간에 따라서 플라즈마 종들 사이의 상호 충돌을 나타낸다. 이온화, 재결합과 같은 것이 있고, 어떻게 이들이 전체 분산에 영향을 미치는지 보여준다.  

#### <center> 볼츠만 방정식의 0차, 2차 모멘트의 정의<center>
    
    
    





적분은 속도 좌표에 대해 모든 공간에서 수행된다. 1차 모멘트의 솔루션과 함께 이 두 모멘트는 모멘텀 방정식으로도 불린다. 입자 종의 에너지 방정식과 연속성을 얻기 위해 풀어야 하는 식이다(입자 질라야 $m$에 대해).


이 식에서의 종($s$)들은 전자와 양이온으로만 한정된다. 여기서는 에너지 방정식의 결과만을 제시한다.
