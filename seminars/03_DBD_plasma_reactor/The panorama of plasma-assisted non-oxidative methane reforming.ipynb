# The panorama of plasma-assisted non-oxidative methane reforming

# Abstract

- 플라즈마의 분류와 플라즈마 화학에 대한 설명  


- 2010~2016년 사이의 non-thermal plasma를 사용한 메탄 커플링 반응에 대한 논문에 대한 심도있는 리뷰  


- 그 중 가장 좋은 결과를 낸 non-thermal plasma 기술을 소개  


- 플라즈마를 사용한 메탄 커플링 반응과 열에너지를 이용한 메탄 커플링 반응의 에너지 비용을 비교

# 1. Introduction

다양한 메탄 개질 기술을 사용한 다양한 생성물의 분포가 연구되어왔다. 이 중 비산화 메탄 커플링 반응, 열분해, 부분산화, 건조와 스팀 개질이 가장 상업적인 가치가 크기 떄문에 주목받고 있다.

- #### 비 산화 메탄 커플링 반응  


두 분자가 직접 합성되어 고온(1000'C 이상)하에서 아세틸렌과 에틸렌을 형성한다.  


$CH_{4} \rightarrow  \frac{1}{2} C_{2}H_{2} + \frac{3}{2}H_{2}$ 　　　　　$\Delta H = 193.5 KJ/mol$  
$CH_{4} \rightarrow  \frac{1}{2} C_{2}H_{4} + H_{2}$ 　　　　　　 $\Delta H = 101 KJ/mol$

<br>

- #### 메탄 열분해  

수소와 탄소를 생성한다.  

$ CH_{4} \rightarrow C + 2H_{2}$ 　　　　　$\Delta H = 74.8 KJ/mol$

<br>
- #### 산소, 이산화탄소, 합성가스 증기를 통한 개질

$ CH_{4} + \frac{1}{2}O_{2} \rightarrow CO + 2H_{2}$ 　　　　　$\Delta H = -35.6 KJ/mol$  
(부분 산화)  

$CH_{4} + CO_{2} \rightarrow 2CO + 2H_{2}$ 　　　　$ \Delta H = 247.3 KJ/mol$  
(건조 개질)  

$ CH_{4} + H_{2}O \rightarrow CO + 3H_{2}$ 　　　　$\Delta H = 206.3 KJ/mol$  
(스팀 개질)

이러한 공법들이 혼합되어 사용되는 경우도 있다.

플라즈마를 이용한 메탄 개질반응은 열분해 공정의 여러 문제점을 극복할 수 있는 대체공정으로 고려되고 있다. 
저압, mild 온도 (<< 1000'C) 하에서 매우 빠르게 반응종은 반응을 일으킨다. 또한, 빠른 반응속도와 높은 에너지 밀도 때문에 상대적으로 낮은 반응 부피가 필요하다. 마지막으로, 재생가능한 에너지원에서 제공되는 전기에너지를 사용할 경우 $CO_{2}$의 방출을 줄일 수 있다.

# 2. Plasma classification


- #### Thermal plasma  

모든 생성종은 열적 평형이므로 같은 온도이다. 절대온도는 $10^{4}K$정도로 매우 높고, 이러한 조건에서 활성종은 thermal activation이나 electron collision을 통해 희박하게 생성된다. 아크방전은 열플라즈마의 예시로 금속공학이나 아세틸렌 합성과 같은 고온을 필요로 하는 공정에 널리 사용된다.


- #### Non-thermal plasma

사용 플라즈마의 종류에 따라 온도가 다르다. non-thermal plasma 내의 종의 온도 크기는 $ T_{electron} >> T _{ion} > T_{neutral`}$ 순이다. DBD, 코로나, nanosecond pulsed discharge가 보편적인 non-thermal 혹은 cold 플라즈마의 예시이다.

- #### Warm plasma

local thermal equilibrium에 도달하지 못하는 플라즈마이고, 중간적인 행동을 보인다. Microwave, radio frequency, glow, spark 방전등이 속한다.

# 3. Plasma chemisty: reaction mechanism

일부 저자들은 Thermal Arc, DBD, MW, Gliding ARC, DC micro-glow와 같은 특정 플라즈마 기술을 구현한 모델을 사용하였고, 일부는 보편적인 조건과 그 결과를 다른 방전으로 확대시켰다.

### 3.1 Plasma chemistry for thermal plasma

아크방전에서 중성종과 이온화종은 열적 평형 상태에 있다. 고온에서 반응성이 있는 중성화 종이 더 많다. 따라서 메탄의 탈수소화가 진행되고 재결합하여 에탄을 형성하고 탈수소화 하여 에틸렌, 아세틸렌을 형성한다.



### 3.2 Plasma chemistry for non-thermal plasma

낮은 전자에너지에서는 vibrational excitation, 중간 크기의 전자에너지에서는 분자 분해, 높은 전자에너지에서는 분자 이온화가 일어난다. 비산화 메탄 커플링 반응의 주요 전자 영향 반응이다.




희박한 수의 전자만이 이온화 되지만 이들은 방전을 유지하므로 중요하다.

- Nozaki 등

낮은 E/N (전기장 세기/ 분자 밀도)에서는 메탄의 탈수소화 반응이 일어나 $CH_{3}$ 라디칼이 대부분이 된다. 높은 E/N에서는 $ CH_{2}, CH, C$가 주요 하다. 따라서 생성물의 분산이 다르다.

- Yang 등

주입한 에너지가 낮을 시 전자 농도와 에너지가 낮고, 에탄이 주요한 생성물이었다. 높은 에너지 주입의 경우 전자 농도와 에너지가 높고 아세틸렌이 주 생성물이 된다.  



<br>
초기 라디칼이 생성 된 이후 반응 메커니즘은 라디칼 결합 반응에 따라 결정된다. 라디칼 결합 반응은 주요하게 가스온도에 따라 결정된다. (분명하지는 않지만 critical 하다)

# 4. Plasma-assisted methane coupling

- 중요 인자는 전압, 공급 유속(feed flow rate), 펄스 반복 횟수이다. 


- 전압이 높으면 강한 방전이 일어나고, 이는 높은 메탄 전화율을 야기한다. 메탄의 분해율도 높아져 $C_{2}$탄화수소가 주된 생성물이 된다.


- 공급 유량이 크면 잔류 시간과 비에너지 공급(specific energy input,SEI)가 작아지므로 낮은 전화율을 야기한다.


- 10KHz 이상에서 Frequency는 더 높은 전화율이 얻어진다.


- 전극의 형상도 영향을 미친다. 뾰족한 끝은 전기장을 강화하고, 짧은 방전 간격(short discharge gap)은 전자 밀도를 증가시킨다. 두 경우 모두 메탄 전화율을 상승시킨다.


- 불활성 혹은 반응성 가스를 메탄 전화율 혹은 특정 생성물의 선택도를 높이기 위해 메탄과 함께 공급하는 경우도 있다.

- 플라즈마와 촉매를 통합하여 사용하는 경우도 있다.

### 4.1 Dielectric barrier discharge (DBD)

전압, 공급 유속, frequency, 펄스 지속을 달리하면 플라즈마의 특성이 변한다.

- Lu and Li  
대기압, 상온의 DBD에서 non-thermal plasma를 사용하여 메탄 전화 공정 연구  
적용 전압이 증가하면 메탄 전화는 증가하고, $C_{2}$선택도는 감소한다.  
공급 유속이 증가하면 메탄 전화는 감소하고, $C_{2}$선택도는 증가한다.  


- Wang 등  
상압하의 DBD와 co-axial micro-DBD 반응기에서 순수 메탄의 높은 탄화수소로의 직접 변환  
Micro-DBD가 더 높은 성능을 냈다. 가장 높은 메탄 전화율은 25.1%, $C_{2},C_{3}$선택도는 80.27% 이다.


- Xu and Tu

메탄의 잔류시간이 메탄 전화율과 선택도에 가장 큰 영향을 미치는 인자이다. 잔류시간을 증가시키면 선형적으로 메탄 전화율이 상승하고, $C_{2}$ 선택도는 상당히 감소한다.


- Liu 등

공급 유량의 증가는 수소 수율과 메탄 전화율을 감소시키나 $C_{S}$ 선택도는 증가시킨다.

짧은 펄스를 여러번 공급하는 것이 더 효율이 높다.  
여러번의 좁은 펄스를 높은 전압 공급원으로 사용할 수 있다.

### 4.1.2 Analysis of the performance





위 그림은 비에너지에 대한 메탄전화율, 아래 그림은 비 에너지에 대한 에탄선택도를 도시한 그래프이다.

- 비에너지의 증가는 높은 전화율과 에탄의 낮은 선택도를 야기한다.  
- 비활성 가스는 성능에 영향이 약하고, 높은 가격을 고려하면 상업화의 해결책으로 사용할 수 없다.  
- 촉매의 사용은 방전의 성격을 바꾸고 성능에 영향을 준다. 촉매 본질에 따라 촉매의 영향이 결정된다. 일반적으로 에탄 선택도를 상승시키지만 메탄전화율에는 일정한 경향을 보이지 않는다.
- 따라서 특정 플라즈마 방전에 특화된 촉매의 개발이 필요하다.

## 4.2 Gliding arc (GA)

클라이딩 아크는 두 동일 평면상의 갈라진 전극에서 생성되기 시작하고 가스 흐름이 플라즈마를 전극의 끝으로 몰고간다. 글라이딩 아크의 길이가 임계값을 넘으면, 열손식이 에너지 공급을 초과하여 플라즈마는 열적 평형상태를 유지하지 못하고, 비평형 플라즈마로 변한다.  


- Dae Hoon 등  
rotating 아크 반응기에서 아크 길이가 $C_{2}$ 화합물의 선택도에 미치는 영향 연구  
원통형 반응기에서의 아크 길이가 길었지만 선택도와의 연관성의 결론은 찾지 못했다.

- 촉매의 사용은 촉매의 특성과 위치에 따라 선택도와 전화율 모두를 증가시킨다.

# 5. Comparison among all plasma techiques used for methane coupling and benchmarking with thermal energy-driven methane coupling  



#### 1. 주 생성물만을 고려한 평가  

$C_{2}$화합물, 수소, 탄소가 주 생성물로 얻어진다.

주 생성물들의 생성 에너지 장벽을 나타낸 그래프이다.  

$SEI = 1345 \times Power [W]/Flow CH_{4} [mlmin^{-1}]$ [KJ/mol]  

$ SER(필요 비 에너지) = SEI/Conversion $ [KJ/mol]  
한 메탄 분자가 완전히 전화하기위한 필요 에너지  

$ \eta $% $ = 100 \times \Delta H_{r}/SER $  

ER = 2 $\times SEI/(Conversion \times Main Product Selectivity)$ [KJ/mol]  
한 아세틸렌 분자를 형성하기 위한 에너지


1. 고에너지 방전에서 아세틸렌이 주생성물인 경우(GA, MW, spark)  
2. 방전의 특정 조건하에서 탄소가 얻어지는 경우  
3. DBD와 같이 에탄이 주생성물인 경우

1, 2 번의 경우 반응 엔탈피와 효율이 약간 낮게 측정되었을 수 있고 3 번의 경우 약간 높게 측정되었을 수 있다. ($C_{3}$ 이상의 탄화수소가 각각 불포화, 포화이기 때문)



- Thermal arc plasma(Huel process,Dupont process)  
아세틸렌과 탄소를 생성에 특화되어 개발되었다.  
캐리어가스로 수소를 높은 농도로 공급할 시 아세틸렌이 ~76% 까지 생성된다.  
공급 가스를 500'C까지 예열하면 효율은 46% 까지 상승하고, 에너지 비용은(ER) 25% 정도 낮출 수 있다.


- Non-thermal discharge at lower temperature  
DBD가 가장 낮은 효율을 보인다. 그 이유는 전기장 세기가 낮으므로 메탄의 메탄의 분해보다 메탄의 진동운동에 에너지가 쓰이는 경우가 많기 때문이다.(들뜬 메탄종의 밀도가 전자 밀도보다 150배 크다.) 결과적으로 1% 근처의 효율을 나타낸다.  
효율을 높이기 위해 나노초 펄스 전압으로 이온 전류를 생성해 원치않는 에너지 손실을 막는 방법이 고안되었다.  가장 효과적인 방법은 방전구역에 tailored 촉매를 넣는 것이다. 열손실의 일부를 촉매반응에 효과적으로 이용할 수 있고 효율을 상승시킬 수 있다.


- Other plasma techniques  
높은 에너지밀도를 갖는 경우 높은 효율을 갖는다.  
코로나방전과 MW 방전은 대기압하 4~6%정도의 효율을 갖는다.  
몇몇 펄스 코로나 방전과 GA는 15~ 20% 효율에 도달한다.  
마이크로 방전에서는 높은 local 가스 온도가 에탄과 메탄을 활성시킬 수 있으므로 높은 효율을 얻을 수 있다.  
전압이나 진동수를 높여 방전 내 축적 에너지를 높이면 탄소 축적을 저해할 수 있고 생성물로부터 열에너지 회복을 최대화할 수 있다.  


- 열플라즈마와 비교가 가능한 경우  
 1. Babarisky등  
 80 mbar의 MW플라즈마 반응기에서 실시. 61%의 효율을 냈고 ER은 604KJ/mol$_{C2H2}$이다. 고압에서는 방전은 더 warmer해지고 아세틸렌의 탄소화를 촉진하며 열손실은 늘어난다.  
 그 결과 효율은 6.9%로 줄어든다. 저자는 평형에서 멀리 떨어진 저압에서 작동하는 것이 좋다고 결론지었다.  
 
 
 2. Yao 와 Suzuki  
 대기압 하 펄스 방전을 사용. 효율은 46%, ER은 860KJ/mol$_{C2H2}$이고 열플라즈마와 비교할만 하다. 분해의 진동 channel이 높은 threshold 에너지, 낮은 cross section을 가진 다른 전자 효과 분해루트보다 지배적이기 때문에 높은 효율을 나타낸다. 하지만 업스케일이 현재 힘들다.  

  

SEI에 대한 메탄전화율과 아세틸렌 수율을 나타낸 그래프이다.  

- 위의 표와 그래프의 결과로 DBD와 코로나 방전은 매우 낮은 메탄 전화율과 낮은 아세틸렌 수율을 나타낸다.  
- GA는 DBD나 코로나 방전보다는 우수하지만 여전히 thermodynamic equilibrium에 미치지 못한다.  
- 펄스방전과 아크방전은 더 나은 성능을 보이고 thermodynamic equilibrium과 근접하다.

# 6. Outlook

- 비평형 전기 방전, 전기 에너지의 분자 분해로의 직접 채널링은 열분해 공정의 대책이 될 수 있다.  
- Cold 플라즈마의 장점은 낮은 작업온도, 낮은 열관성, 작업 조건의 변동에 빠르게 반응하는 점이 있지만 높은 효율을 얻는 것이 과제로 남아있다.  
- 플라즈마 특성에 영향을 주는 인자들을 조절함으로써 효율을 상승시킬 수 있을 것이다. (ex. 소멸되는 에너지의 사용)  
- 플라즈마와 촉매를 통합하려는 노력이 진행되고 있고, 둘은 시너지 효과를 나타내는 것으로 관측된다. 하지만 과학적, 기술적으로 본질적인 과제(코크형성, 표면반응 등)가 남아있다,  

# 7. Conclusions

Non-thermal 플라즈마는 중간(cold plasma) 혹은 높은(warm plasma) 가스온도를 갖는다. non-thermal 플라즈마에서 가스온도는 반응의 활성과 반응의 재결합에 영향을 주기 때문에 생성물 형성과 생성물 분산에 매우 중요하다. 너무 높은 온도는 열손실을 줄이고 공정효율을 높이기 위해 피해야한다.  

펄스방전은 펄스 진동수를 통해 가스 온도를 조절함으로써 방전 기간의 화학반응 뿐아니라 연속 방전 사이의 switch-off 기간을 제어할 수 있다. 특히 높은 효율에는 나노초 펄스 전압과 높은 진동수가 추천된다.  

촉매의 사용은 한편으로는 생성물 선택도를 상승시키지만 다른 한편으로는 전기장을 왜곡하여 반응기 성능을 감소시킨다. 따라서 tailored 촉매가 필요하다.  

DBD와 코로나 방전은 낮은 반응기 성능을 보인다. GA는 조금 더 나은 성능을 보인다. 펄스 스파크와 MW 방전은 가장 높은 성능을 보이며 열아크 방전에 근접하다.  

공정 효율과 에너지 필요량을 고려할 시 몇몇 non-thermal 플라즈마(펄스 방전)는 열분해 공정과 견줄 수 있다. 현재는 업스케일링의 문제가 존재하지만 재생가능한 에너지의 활용이 증가한다면 미래에 플라즈마는 메탄 valorization의 귀중한 기술이 될 것이다.
