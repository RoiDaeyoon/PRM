# Dynamics of dielectric barrier discharges in coplaner arrangement

# Abstract  

- 동편면상의 방전 배리어 배치에서 방전 channel의 발달을 수치적으로 조사하였다.  
- 산소안에서 장세기의 시간, 공간적 분산, 입자, 에너지 밀도와 같은 행동들을 자세히 묘사한다.  
- 스트리머의 발달은 주로 광전자방출로 일어난다. 
- 음극층은 음극에서 직접나온 스트리머가 유전 표면을 건드리는 위치 근처에서 나타난다.  
- 이온 붕괴로 인한 이차전자는 중요하고, 음극층의 변수들은 normal glow discharge의 것과 가깝다.  
- 전하이동과 에너지 방출은 음극 스트리머의 발달의 결과인 유전체 표면에서 나타나는 방전의 전도성 channel에서 발생한다.
- 전도성 channel에서 장의 세기는 70~100Td(산소, 공기속)으로 거의 일정하다.

# 1. Introduction

보편적인 DBD와 달리, 유전 평면위에 DBD을 발달시키는 장치가 있다. 만약 유전층이 유전판위에 있는 전극을 덮고 있다면, 방전channel은 각 쌍의 전극 사이의 유전층을 따라 발달한다.



보통 coplanar discharge(CD)의 전극 거리는 100um정도이고, 유전층의 두께는 10~60um 정도이다.  

CD는 PDP에서 많이 쓰이고, 오존의 합성에서의 전망이 밝다.




방전은 유전 표면에의 전극 위치에서 생성되고 종결되는 분명한 평형 channel을 포함하고있다. 채널사이의 평균거리는 내부전극 거리에 영향을 받는다. 채널은 동시에 나타나고, 동시에 발달한다. 


이 논문의 목적은 coplanar arranement에서 방전의 경향을 수치적으로 조사한 결과를 요약하는 것이다. Dinymis중 가장 중요한 것중 하나는 전기장에서 charged 입자, 내부 원자와 분자의 상태에의 에너지 전달이다. 오존합성이 이 목적을 적용하기 편리한 예시로 선정되었다.  
이 논문에서는 하나의 채널의 dynamics만을 살펴본다.


# 2. Numerical model

DBD모델링은 방전 구역에서 충전된 입자의 dynimics와 전기장의 시간에 따른 변화를 계산하는 것이다. 채널의 폭은 다른 채널과의 거리와 비교했을 때 작다. 전극 거리를 100um이하로 줄이는것과 채널간의 거리보다 작게하는 것은 실질적으로 동일하다. 따라서 2차원 접근법이 선택되었다.  

이 경우 z축에서의 방전은 동일한 것으로 본다. 만약 방전의 채널 구조도 중요하다면, 채널의 폭도 고려되어야 한다. 다음은 방전 구역의 그림이다.  




## 2.1. Transport equation, initial and boundry conditions

charged, 중성 입자 밀도의 역학은 볼츠만 방정식으로 묘사된다. 볼츠만 식의 결과와 local field approximation(LFA)로 얻은 결과는 큰 차이가 없다. 이는 모든 입자의 공간적, 시간적 분산을 모샤한다. 

<center> $\frac{\partial{n_{i}}}{\partial{t}}+\mathbf{div}(\overrightarrow{j_{i}})=S_{i}$</center>  
<center>$\overrightarrow{j_{i}}=D_{i}\cdot \mathbf{grad} n_{i} + \overrightarrow{v_{i}}\cdot n_{i}$</center>

n은 charged입자 밀도의 수, j는 전류 밀도, D는 확산도, v는 플라즈마의 iqjs 구성성분의 속도이고, S는 소스텀으로 이온화, 흡탈착 공정등을 포함한다.


가스 공간내에서 integration region을 제한하는 유전 표면에는 2개의 경계조건이 존재한다. (external boundray) 유전 표면은 흡착 입자에 의해 취해지고, 2차전자로써 떨어져 나간다. 2차전자 플럭스의 결과는 다음과 같다.  
<center>$n_{s}\cdot v_{e} = \gamma _{+}\cdot n_{+}\cdot v_{+} + \gamma_{ph}\cdot \phi$</center>

$\phi$는 표면에서의 전자 플럭스 밀도이고 이온화도에 비례하고 전체 방전 부피에서 모두 더함으로써 결정된다. $\gamma_{ph}, \gamma _{+}$는 방출 coefficient로  0.001~0.3 까지 모델링동안 변화한다.

외부 경계층은 열려있다. 이 경계에 도달하는 입자는 integration 지역을 떠나는 것이다.  
(A) 초기 전자-이온쌍은 유전 표면 근처의 층에 균질하게 분산되어있다.  
(B) 가스 구역에서 알짜 charge는 0이다.  
(C) 방전 표면의 초기 전하밀도는 0이다. 여러 세트의 방전이 고려된다면 초기 전하밀도는 이전의 잔류 전하밀도에 의해 결정될 것이다. 

## 2.2. Field calculation  

가스 부피와 유전 평면의 다른 유전율하에서의 CD arrangement의 field configuration을 찾기 위해서 다음의 Poisson 방정식을 풀어야 한다.  
<center>$\mathbf{div(\overrightarrow{E})}  = \frac{1}{\epsilon _{0}\cdot\epsilon_{i}}\cdot \rho_{i}$</center>  

$\epsilon _{0}$는 상대 유전율, $\rho_{i}$는 전체 전하밀이다. 방전 표면에서의 포텐셜과 intefration region의 외부 경계를 알아야만 한다.  
경계 포텐셜을 결정하기 위해 영상법을 사용한다. 각각의 존재하는 전하들이 영상전하를 가지고 있다면 방전은 고려하지 않아도 된다. 전하의 총합(전극 표면전하, 방전 표면전하, space 전하)을 알면 특히 유전 표면에서의 정확한 포텐셜 분산을 결정할 수 있다.  

방전 arrangement는 여러개의 임피던스(전압과 전류비)를 가진 전기 회로의 요소이다. 방정식의 계의 각각의 해를 구하기 위해 두가지의 asymptotic 접근을 사용한다.  

회로가 매우 큰 임피던스를 갖는다면 전극의 전하는 breakdown동안 변하지 않는다.(Q=const) 하지만 전극의 전압은 감소한다.  
임피던스가 거의 0이라면, 전극에서의 전압은 변하지 않는다.(V=const)

## 3.1. Electric field distribution  

2100볼트 하에서의 유전 표면위의 초기 장의 세기 분산은 다음 그림과 같다.  


왼쪽의 양극, 오른쪽의 음극 모두 양의 값을 갖는다.이는 왼쪽에서 오른쪽으로의 전자 유동이 일어남을 의미한다.   


방전을 개시하기위해 seed 전자가 방전 구역으로 주입된다. 이 전자들은 양극으로 이동한다. 분해등으로 이들은 증식하고 특정 입자 밀도 이상에서 장의 왜곡이 나타난다.(1.3ns) 최대 부분은 음극 쪽으로 50~70um 만큼 이동한다.이 propagation 속도는 약 $10^{7}cm/s$로 전자 전자 유동속도와 비슷하다. 스트리머는 전자 플럭스의 반대 방향으로 이동한다.(cathode directed streamer)  
양극쪽의 위에도 장 세기의 최댓값이 나타난다.(1.3ns) anode directed streamer으로 보인다. 두 스트리머의 두께는 모두 20~30um이다. 이러한 동시움직임이 초기 장의 분산을 부드럽게 만든다.   
스트리머의 빠른 이동동안 두 최댓값 사이의 간격은 늘어나고 150~200um의 길이가 된다.


y축의 에너지 우측 부분에서 깊은 최소 장 세기값(-900Td)을 보인다. 1.3ns에 두께 20um, 너비는 50~60um이지만 23.9ns에는 150um까지 늘어난다. 음극층은 2차전자의 효율적인 sourcec가 되고, 이로 인하여 두 스트리머의 전파가 가능하다. 음극층이 형성된 이후 전자방출은 덜 중요하다.  
양극에서 장의 세기는 0이지만 전자는 계속 움직여 소멸 기간에 양극의 끝부분에 도달한다.


소멸단계에는 전자 플럭스가 감소하고 음극에 양의 전하가 축적되어 음극층의 장세기는 감소한다. 음극층의 너비는 계속 넓어지며 소멸 단계에 가장 끝부분까지 도달한다.  

스트리머가 성장하고 음극층이 형성되는 방전단계는 5ns정도 까지이다. 전자단계 이후의 소멸 단계동안은 모든 방전 과정은 평균 장 세기의 하락으로 인해 감소한다. 이 시점부터 이온의 움직임이 방전의 작용을 결정하고, 장 세기의 분산을 더 부드럽게 만든다.

## 3.2. Dynimics of charged particles  

전하 이동은 최대 전자 밀도와 함께 매우 전도성이 있는 방전 채널을 형성한다.


채널은 장의 세기가 안정화 된 곳에서 발달한다. 125um 까지 길어지고, 두께는 20~25um이다. 전자밀도는 최대치에 도달하고 종말단계에서 연속적으로 감소한다. 채널의 알짜 전하는 0이고, 유사 중성이다. 


전자 생성율은 흡착율과 전자의 이동에 의해 균형이 이루어진다. 채널에서 전자밀도가 떨어지면 장 세기는 급히 오른다. 이는 이온화율의 상승과 흡착의 감소를 동반하여 전자밀도를 상승시키고, 그 역도 성립한다. 따라서 장의 세기가 0까지 떨어지지 않고 70~100Td를 유지한다.



표면 전하 밀도는 연속적으로 상승한다.  


전자 단계는 주로 유전 표면에 전자가 모인다. 일정 수준까지 표면 전하밀도가 상승하면 표면을 넘어 유전 표면의 더 먼 곳까지 도달해 전하 spot의 크기를 증가시킨다. 이온 단계와 종말 단계에서 이온의 움직임은 표면 전하 분산을 부드럽게 만들고, 더 대칭적인 모양을 갖게 한다.  
전하밀도는 15~30nC/Cm$_{2}$을 넘지 못하다가 전자 방충이 더해지면 전하의 이동이 5~6배 빨라져 전하 밀도에 영향을 준다.

## 3.3. Ion phase  

전자단계가 지나면 전자의 number density가 소멸단계의 규모정도로 줄어든다. 방전 표면에 모인 전하가 방전 구역의 장 세기를 감소시킨다.  또한, 들뜸, 분해 과정이 종결된다. 많은 수의 이온이 방전 구역을 떠나고 이 유동은 전기장을 더 감소시키고 표면 전하를 상승시킨다. 이온단계 이후 다음 방전 펄스의 초기 상태가 주어진다. 이온 유동은 또한 에너지 손실을 일으켜 효율을 감소시킨다.  

이온단계는 1um이하로 지속된다. 이 단계가 지나면 표면전하밀도는 대칭적으로 되고, 방전표면의 charged 지역은 전극의 끝까지 넓어진다. 양극 사이의 장 세기는 0에 도달하고 음극은 사라진다.  


이온 유동은 0.5um 정도에 끝난다. 일부 이온이 유전 표면에 도달하고 그중 일부는 방전 볼륨을 떠난다. 방전 볼륨은 실질적으로 중성이다. 이온은 장세기가 약해져서 유동이 불가능하고 재결합을 통해서만 중성화가 가능하다. 이 단계가 지나면 방전 구역은 전하된 입자에 자유롭고 장 세기는 0이 된다.

## 3.4. Discharge current 

전류는 전하의 변화이다. 이 값은 V=const 접근으로 찾을 수 있다. Q=const인 경우는 방전은 일어나지만 외부 회로에 전류가 흐르지 않는 경우이다. 이런 경우 effective 전류는 가스 구역의 전하 플럭스로 정의된다. Sato-Morrow 방정식을 사용하여 계산된다.  

V=const인 경우 2.2절의 방법을 사용해도 Sato-Morrow 방정식과 같은 결과를 얻을 수 있다.


전류의 형태는 급격히 상승하였다가 상대적으로 천천히 감소한다. 전류의 상승은 방전표면에서의 음극 층 신장의 속도로 결정된다. 이 과정에서 음극층의 전류 밀도는 일정하다.  

방전 기간은 2~4ns 정도로 상대적으로 짧다. 이는 가스 gap에서의 배열된 것에서 측정한 것보다 몇배 짧은 시간이다. 전류 모양은 외부 회로에 적은 영향만을 받는다. V가 일정한 경우가 조금 더 넓고 진폭이 40% 정도 높다.

# 4. Energy release and efficiency  

## 4.1. Energy release in the discharge volume

전류의 흐름과 함께 에너지는 방출되기 시작하고 charged 입자 밀도가 급격히 감소할때 멈춘다. 에너지 방출은 전자와 이온의 유동에 의해 결정된다.  
에너지 밀도 $w$는 local 장 세기와 유동속도의 스칼라곱의 총 합을 시간으로 적분한 것이다.  
<center>$w=e_{0}\int (\Sigma \left | \overrightarrow{E}\cdot\overrightarrow{v_{i}} \right |\cdot n_{i})dt$  </center>

$e_{0}$는 기초 전하이고 나머지 변수는 앞의 것과 동일하다. 모든 charged 입자는 에너지 소비에 기여한다.  

 

첫 에너지 방출은 양극에서 스트리머의 성장과 함께 나타난다. 음극 directed 스트리머의 뒷부분에 매우 에너지 밀도가 높은 지역이 나타나고, 스트리머가 움직이며 더 커진다. 최댓값은 20um까지의 높이를 가진다. 에너지 밀도의 평균 값은 15mJ$cm^{-3}$이다.
이러한 에너지의 일부는 이용의 목적으로 사용할 수 있다.


산소원자 밀도의 분산을 예시로 보여준다. 위 그림의 에너지 밀도와 비슷한 모습을 보인다.



대부분의 산소 원자는 방전 채널에서 생성된다. 비록 산소원자의 생성율과 장의 세기는은 스트리머 헤드 부분에서 가장 크지만 대부분의 산소 원자의 수율은 상대적으로 전기장 세기가 약한 채널에서 얻어진다. 헤드부분의 시 공간적 부피가 작은 것이 이유이다.

## 4.2. Efficiency of electron collision processes  

이온단계가 지마고 필드에서 입자로의 에너지 이동이 완료되고, 에너지 소비의 최종값, 들뜬, 이온화된, 분해된 입자의 수가 고정된다. 이러한 값들은 공정의 효율을 결정한다.
  

산소 원자는 3.2ns 정도 까지 증가한다. 10~15ns에 시작되는 이온 단계에서 에너지 소비는 이어지지만 산소 분해는 멈춘다. 전자단계 동안 에너지 소비는 낮은데 이는 이온유동으로 부터의 손실이 무시가능하기 때문이다. 그러나 이온 단계가 지나면 에너지 손실이 크게 증가한다.  

오존 합성의 효율은 effective 장 세기에 의존한다. 최대 효율은 200Td 근처에서 형성된다. 하지만 효율의 최종값은 100Td 에서의 값과 비슷하다.  

40Torr정도의 낮은 압력에서 short UV펄스로 전자가 생성된다. 그림 14에서 높은 E/n 값은 breakdown이 없었다고 생각할 수 있다. 이러한 조건에서 charged 입자의 밀도는 균일한 장의 분산을 왜곡하기에는 너무 약하다. 또한 산소 분해는 거의 일정한 장에서 일어난다.  

외부 회로의 변수는 방전 역학과 에너지 소멸과 들뜸, 분해 공정에 영향을 준다. 
  
V=const, Q=const 의 경우를 고려하면 몇몇 변수가 다르다. 에너지 값, 이동한 전하, 오존 생성의 비에너지 소모가 방전 발전의 3 단계로 제시되어 있다.  

전자 단계의 전류 펄스의 지속은 두 경우 모두 4~6ns이다. V=const의 경우 이동된 전하와 에너지 소비는 Q=const의 경우보다 높다. 이는 전압 하강을 일으킨다.  
Q=const의 경우 더 일찍 일어나고 에너지 소비가 15~20% 적다.
