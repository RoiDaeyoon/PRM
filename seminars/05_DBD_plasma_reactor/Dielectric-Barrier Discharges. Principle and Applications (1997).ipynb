# Dielectric-Barrier Discharges. Principle and Applications (1997)

# Abstract

- DBD는 넓은 온도, 압력 범위에서 작동될 수 있는 비평형 방전이다.  
- 대기압 정도에서도 잠깐 동안의 microdischarge는 고압에서의 글로우방전의 특성을 가져 배경 기체의 원자, 분자를 들뜨거나 분해시킬 수 있다.  
- large-scale 오존 생성 공정, 표면 처리, 오염 조절 등의 적용에 관해서 논의한다.

# 1. Introduction

DBD는 오존생성에 주로 적용되고, 오염 조절, 표면 처리 공정에도 좋은 효과를 나타낼 것으로 보인다. ac PDP는 가장 좋은 효과를 나타낼 것이라고 생각된다.


DBD의 가장 중요한 특성은 비평형 플라즈마 상태라는 점이다. 지리적인 표면 형태, 작용 미디엄, 작용 파라미터는 전례없을 정도로 flexible 하다. lab-scale에서 최적화 된 파라미터는 쉽게 산업에 적용 가능할 정도로 스케일 업이 가능하다. 

# 2. Discharge structure and microdischarge properties




DBD는 금속 전극과 방전 공간사이 전류 path에 단열 층이 존재한다. 금속전극에 의해 생성된  강한 전기장으로 인해 유전체의 표면에 방전이 시작된다. 유전체의 존재로 직류는 사용할 수 없다. 일반적으로 DBD는 선주파수, 마이크로웨이브 주파수 사이의 500Hz ~ 500kHz 사이에서 작동한다.

유전체로 나뉘어진 가스 공간은 금속 전극 사이에 있는 것과 같은 breakdown 전압을 갖는다. 

## 2.1 Microcharge properties

DBD의 특성 중 하나는 대기압 하에서 나타나는 breakdown의 대부분의 기체가 수많은 독립된 전류 필라멘트나 마이크로방전의 형태로 나타난다는 점이다. 사진2는 1mm air gap으로 투명한 전극에 통과시킨 사진이다. 그림 3은 각각의 마이크로방전의 자취를 찍은 사진이다.

마이크로방전의 필라멘트는 일시적은 고압의 글로우 방전과 특성이 비슷하다. 유전체 위에 생긴 charge로 인한 마이크로방전의 위치의 장은 수 ns 만에 줄어들고 전류 흐름을 종결시킨다. transported charge는 gap의 크기와 유전율에 비례하지만 압력에는 영향을 받지 않는다. 전류 밀도는 100~1000 A$cm^{-2}$까지 도달한다. 습도는 마이크로 방전의 세기를 증가시키는 경향이 있지만 음극에 UV를 조사하는 것은 감소시키는 경향이 있다. 

유전 베리어는 charge의 총량과 에너지 축적을 제한하고, 전체 적극으로 마이크로 방전을 분산시킨다. 외부 전압이 상승함에 따라 추가적인 마이크로 방전이 새로운 위치에 발생한다. 이는 유전체의 잔류 전하가 이미 마이크로 방전이 일어났던 위치의 전기장을 약하게 하기 때문이다. 반면 전압이 약해지면 이미 존재했던 위치에 마이크로 방전이 일어난다.

따라서 고압, 낮은 진동수의 작동이 마이크로 방전을 확산시키는 경향이 있다.

전압 모양, 주파수는 가스의 특성, 압력, 전극 형상에 의해 결정된다. 파워의 상승은 단위시간, 단위 면적당 마이크로방전 생성의 수를 증가시킨다. gap 거리와 파워 밀도가 같으면 scale에 따른 문제가 발생하지 않는다.  

개개의 마이크로방전 물성은 매우 빠른 전압의 상승이 있으면 변할 수 있다. 전압의 상승시간이 마이크로 방전의 지속시간과 비슷하면 동시에 수많은 마이크로 방전이 함께 형성될 것이다. 따라서 방전이 일어날 수 있는 충분한 표면이 부족하다. 이러한 경우 더 약한 마이크로 방전을 야기한다. 

## 2.2 Overall Discharge Characteristic

개별 마이크로 방전의 균일성의 묘사는 전체 방전 특성의 정의를 이끈다. 그림 4의 Lissajous figures의 전압,charge를 기록함으로써 보일 수 있다.  






사인파의 전압을 적용하였다고 가정한다. 초기에, 방전 없이, 전체 정전용량 $C_{tot}$는  전압의 상승에 따라 충전된다. 가스 공간의 Breakdown 전압이 마이크로방전이 시작되는 점까지 도달하면 유전체의 정전용량 $C_{D}$가 충전되기 시작한다. 방전 전압 $U_{Dis}$, 가스 공간의 평균 전압은 일정하게 유지된다. 

이는 최대 전압 $\hat U$에 도달하거나 두번째 반파장에서 반대의 상황이 일어나면 멈춘다. 많은 DBD 적용에서 voltage/charge Lissajous figure는 이상적인 평행사변형꼴이 된다. 기울기로부터 실질적인 정전용량 $C_{tot},C_{D}$가 구해지고, $U_{Min}, \hat U$는 즉시 결정된다. 방전 전압이 일정하다면 voltage/charge Lissajous figure는 적용된 전압의 형태에 독립적이다. 가정에 무관하게 닫힌 영역은 파워에 비례한다. 


적용된 voltage/charge Lissajous figure와 해당하는 gap voltage/charge 의 제공은 방전의 행동과 designing 파워 필요량과 알맞은 회로의 특성에 매우 유용한 재료이다. 

# 3. Modeling dielectric-barrier discharges

## 3.1 Modelling of the reaction kinetics



시작부분의 공간적 기울기와 균일 플라즈마의 취급을 무시하는 것이 종종 편리하다. 짧고 높은 전압 펄스와 전자 빔의 주사가 short-lived 마이크로 방전의 시뮬레이션을 위해 적용된다. 두 경우 모두 볼츠만 방정식을 통해 반응 속도 상수를 구하는 것이 중요하다. 


DBD의 계산을 위해 국부적인 장의 근사가 일반적으로 사용된다. 전자에너지 분산은 전기장과 함께 평형이라고 가정되고, 모든 속도 상수들은 평균 전자 에너지, 혹은 감소된 장의 E/n의 유일한 방정식으로 나타낸다. 


비록 이러한 가정들이 대략적이어도 오존 생성의 최적화에 대해 잘 맞고 널리 사용된다. 대기압 하에서 전기장 내의 전자의 가속은 배경 기체의 수많은 붕괴를 이끌어 10ps 이내에 평형 값에 도달한다. 주목할만한 전압 변화와 그에따른 전기장 변화는 더 작고, 일반적으로 ns 범위이다. 따라서 많은 경우 볼츠만 방정식의 정지된 해를 사용하여 정의된다. 


전자 붕괴에 의한 들뜸과 분해는 매우 빠른 공정이고 자유라디칼 반응을 수반한다. 대부분의 자유라디칼 반응은 매우 빠르게 완료된다. 따라서 많은 경우 최종 생성물은 자유라디칼 반응에 의해서 생성되고, 마이크로방전 필라멘트에서 일어나는 일을 시뮬레이션 하는 것은 좋은 근사치를 제공한다. 


혼합 가스의 많은 마이크로 방전의 작용의 시뮬레이션에서 전자의 운동을 무시하는 첫번째 접근은 may also be to repetitively inject certain concentrations of free radicals과 그리고 이어지는 화학 반응을 계산한다. 

## 3.2 Two- dimensional modelling

마이크로 방전의 형성을 모델링하는 것은 대기압하 가스 breakdown울 계산하는 것과 깊게 연관되어 있다. 방전 성장의 초기 단계에 금속 전극 사이에서의 breakdown과 유전체 벽의 사이에서의 breakdown에는 차이가 있다. over전압이 방전 gap에 대기압 하에서 적용되었을 때, 음극에서 전자 avalanche가 시작되고, 임계 장에 도달하게 된다.




대기압 공기, $SF_{6}$하에서의 스트리머의 전파의 계산이 수행되었다.모든 실험들은 스트리머 헤드 부분에서의 극히 높은 전자 장의 발생(꽤 얇은 전도성의 채널)은 최대 전자 밀도($10^{14}~10^{15} cm^{-3}$)를 형성한다. 전파는 전자 drift 속도보다 더 빠른 속도의 이온화 wave가 원인이다.

# Conclusion

DBD는 조절할 수 있는 고압의 가스에서의 비평형 플라즈마 상태의 설립을 위한 이상적인 물성을 가지고 있다. 대기압 근처에서의 작업은 flue gas 스트림이나 대량의 흐름의 연소에 필수적이다. 많은 적용법에서 전자 붕괴로 인한 자유라디칼이 이온이나 전자 그자체보다 중요하다. 대기압 하 자유라디칼 화학은 매우 빠르고 따라서 대량생산, 높은 destruction rate, 표면에서의 빠른 속도의 처리를 가능하게 한다.
