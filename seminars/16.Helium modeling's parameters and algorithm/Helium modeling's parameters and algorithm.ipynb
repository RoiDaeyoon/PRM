전자 이동도와 electron impact reaction 전자의 반응 속도 상수는 BOLSIG[Bol96] 을 통해 얻어진다(균일한 전기장에서). 위에서 얻은 전자이동도, 반응속도 coefficient는 97%헬륨 , 3% 수소의 혼합 기체내에서 얻은 값이다. 

무거운 입자의 이동도와 확산계수는 헬륨가스에 의해 정해진다. 이온 이동도는 아래 그림에서 나타난다.




이때 입자들의 확산도는 아인슈타인 관계에 의해 다음과 같이 정의된다.  
### $\ \ \ \ \ \ $ $D=\frac{k_{B}T\mu}{e}$

**T를 결정하기 위해 다음 식을 활용한다.**

### $\ \ \ \ \ \ $ (1) 이온 : $k_BT=k_BT_g+\frac{m+m_g}{5m+3m_g}m_g(\mu E)^2$
### $\ \ \ \ \ \ $ (2) 전자 : $k_B T_e=\frac{2}{3}\bar\varepsilon$


여기서의 $T_{g}$값은 room temperature인 약 300K를 사용하고, 전자평균에너지는 위에서 주어진 E 값을 사용한다.

중성입자의 확산인자는 [Gri97]에서 가져온 값이다.

for  $ He^{*},  D\times p = 4.5\times 10^{2}cm^{2}Torrs^{-1}$ 로 사용된다. p는 가스압력이다.

이하 p는 대기압, 약 1atm정도로 사용할 수 있다.

2차 전자 방출 인자는 헬륨 이온에 대해서 0.2를 사용한다. [Mcd64] 표면에서의 전기장에 의한 전자방출인자의 변화는 무시한다. 중성종에 의한 2차 전자 방출은 무시한다. 헬륨이온에 의해 방출된 2차전자의 평균 초기에너지는 5eV로 한다.


15 - 20 eV에서 R4, R5와 같은 준안정 상태 이온화과정에서 생성되는 전자는 BOLSIG에 의해 구해진 인자와 맞지 않는다. 이들의 전자는 $e^{*}$로 따로 취급한다. 이동도와 확산계수는 다음 값을 상수로 사용한다.

#### 이동도 : $\mu \times p = 6.8\times10^{5}cm^{2}TorrV^{-1}s^{-1}$  

#### 확산도 : $D\times p = 6.8\times10^{6}cm^{2}Torrs^{-1}$

# 풀어야하는 방정식

### Main Equation

### (1) $\frac{\partial n_p}{\partial t}+\nabla \cdot \mathbf{\Gamma_p} = S_p$

### $\ \ \ \ \ \ \ \ \ a.\ \mathbf{\Gamma}_p =sgn(q_p)\mu_p \mathbf{E} n_p-D_p\nabla n_p $ 
### $\ \ \ \ \ \ \ \ \ b.\ S_p = \sum_{r} c_{p,r}R_r$, 
### $\ \ \ \ \ \ \ \ \ \ \ \ \  (R_r=kn)$

<br>$n_p$ : density, $\mathbf{\Gamma_p}$: flux, $S_p$: source term

a의 $\mu$는 위에서 구한 값이다. E는 제공하는 전기장의 세기이고, $D_{p}$는 아인슈타인 관계로부터 얻어진다.  
b의 소스텀의 농도항은 수밀도 항으로 변형할 수 있다. 반응식 항의 k값도 위에서 구한 값이다.  
따라서 (1)식은 수밀도에 대한 식으로 정리되고, parameter는 모두 구해진다.




## 볼츠만 연속방정식의 경계조건  

### $\ \ \ \ \ \ $ (1) 전자를 제외한 입자 :  $\mathbf{\Gamma}_p \cdot \mathbf{n} = (2a_p-1)sgn(q_p)\mu_p \mathbf{E}\cdot \mathbf{n}n_p+\frac{1}{2}v_{th,p}n_p$
### $\ \ \ \ \ \ $ (2) 전자 : $\mathbf{\Gamma}_e \cdot \mathbf{n} = -(2a_e-1)\mu_e \mathbf{E}\cdot \mathbf{n}n_e+\frac{1}{2}v_{th,e}n_e-\frac{1}{2}v_{th,e}n_{\gamma}-2(1-a_e)\sum_p\gamma_p\mathbf{\Gamma}_p \cdot \mathbf{n}$  
### $\ \ \ \ \ \ $ (3) 전자 : $n_\gamma=(1-a_e)\frac{1}{\mu_e}\sum_p\gamma_p \left [ (2a_p-1)sgn(q_p)+\frac{1}{2}\sqrt{\frac{8(m_p+m_g)m_g}{\pi (5m_p+3m_g)m_p}} \right ] \mu_pn_p$ 


열속도 $v_{th}=\sqrt{\frac{8k_BT}{\pi m}}$의 계산이 가능하다. 또한 이차 방출계수 $\gamma$는 0.2로 주어져 있다.

### $\ \ \ $ Poisson's Eq'n
### $\ \ \ \ \ \ $(2) $\nabla\cdot(\varepsilon \mathbf{E})=-\nabla\cdot( \varepsilon \nabla V)=\rho$
### $\ \ \ \ \ \ \ \ \ \ \ \ (\rho=\sum_p q_p n_p)$
<br>$V$ : electric potential, $\rho$ : charge density


$\epsilon$은 유전상수 값이고, E는 전기장의 세기이다. 마찬가지로 파라미터는 모두 구해진 상태이다.



## Poisson's Equation의 경계조건

### 1. 외부에서 가해지는 일정한 15kv의 직류 전압  
### 2. 가우스 법칙에 의한 표면 전하 고려  

### $\ \ \ \ \ \ \varepsilon_{wall}\mathbf{E}_{wall}\cdot \mathbf{n} - \varepsilon_{0}\mathbf{E}_{gas}\cdot \mathbf{n} = \sigma$
### $\ \ \ \ \ \ \sigma = \int \mathbf{j}\cdot \mathbf{n}dt$
### $\ \ \ \ \ \ \mathbf{j} = \sum_{p}q_{p}\mathbf{\Gamma}_{p}$  

전극을 통해 흐르는 실질적인 전류

### $\ \ \ \ \ \ I = \int\int_{electrode surface}\left ( \mathbf{j}\cdot \mathbf{n} - \varepsilon_{0}\frac{\partial}{\partial t} \mathbf{E}\cdot \mathbf{n} \right ) d^{2}S$ 

- 반응에서의 평균 전자에너지가 주어졌다. 이 평균 전자에너지에 따른 전자 이동도도 주어졌고, 이온 이동도도 reduced field에 의해 주어졌다.
- 중성입자의 경우 확산도가 주어졌다.
- 2차 전자 방출인자가 주어졌다.
- 준안정상태 이온화 과정의 전자는 따로 취급하고, 상수로 주어졌다.

이렇게 구한 파라미터및 계산한 파라미터를 토대도 계산을 실시할 수 있을 것이라고 판단한다.

# 예상 문제풀이 매커니즘






입자, 전자마다 에너지, 이동도, 확산인자가 다르다. 따라서 각 입자, 전자마다 각각의 볼츠만 방정식을 풀어서 수밀도를 구해야한다.  
이후 푸아송방정식을 통해 E 값을 구할 수 있다.  
이 E 값을 통해 재차 볼츠만 방정식을 풀어준다. 충분한 전기장 변화가 일어날 수 있을만큼 반응이 일어날 시간간격을 준다.  
이 값을 토대로 재차 푸아송 방정식을 통해 E값을 구하고, 이를 반복한다.  

 하지만 초기에는 중성가스밖에 존재하지 않는다. 즉, 입자들의 초기 농도를 모르기 때문에 초기계산이 불가능하다. 이를 위해 초기 입자 농도를 고려해야한다.

푸아송 방정식의 경계조건의 경우, $\Delta E$ 값을 제공하는 것이라고 생각된다. 이는 뉴만 경계조건으로 $\frac{\partial E}{\partial n}$의 형태로 제공될 것이라고 예상하였다.
